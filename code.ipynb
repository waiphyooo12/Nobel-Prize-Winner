```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

nobel = pd.read_csv("nobel.csv")
print(nobel.head())
```

```python
#What is the most commonly awarded gender and birth country?

#Store your answers as string variables top_gender and top_country.

print(nobel.columns)
top_gender = nobel["sex"].value_counts().index[0]
print(top_gender)

top_country = nobel["birth_country"].value_counts().index[0]
print(top_country)
```

```python
#Which decade had the highest ratio of US-born Nobel Prize winners to total winners 
#in all categories?

#Store this as an integer called max_decade_usa.
nobel["decade"] =(np.floor(nobel["year"]/10)*10).astype(int)
nobel["us_winner"] = nobel["birth_country"] == "United States of America"
print(nobel["decade"].head())
print(nobel["us_winner"].head())
sns.relplot(data=nobel ,x="decade", y ="us_winner",kind="line")
plt.show()
decade_us = nobel.groupby("decade")["us_winner"].mean()
max_decade_usa = decade_us.sort_values(ascending= False).index[0]
print(max_decade_usa)

```

Which decade and Nobel Prize category combination had the highest proportion of female laureates?

Store this as a dictionary called max_female_dict where the decade is the key and the category is the value. There should only be one key:value pair.

```python
nobel["female_winner"] = nobel["sex"] == "Female"
print(nobel[["female_winner"]])
groupped_female = nobel.groupby(['decade' , 'category'] , as_index=False)["female_winner"].mean()
print(groupped_female["female_winner"].max())
max_female=groupped_female[groupped_female["female_winner"] ==groupped_female["female_winner"].max() ]
print(max_female)
max_female_dict = {max_female['decade'].values[0]: max_female['category'].values[0]}
print(max_female_dict)
```
